```python
conn = sqlite3.connect('C:/Users/kenbun/Desktop/jupyter_workspace/db作成/serach_index.db')
cur = conn.cursor()
```

```python
keyword = '大分'

sql = f"""
SELECT path, width, height FROM ocr_text
INNER JOIN headline_paths ON ocr_text.id = headline_paths.id
INNER JOIN headline_size ON ocr_text.id = headline_size.id
WHERE headline_text LIKE '%{keyword}%';
"""
```

```python
import json
```

```python
cur.execute(sql)
test_search_result = cur.fetchall()
```

```python
%%time
result_list = []
for path, width, height in test_search_result:
    result_list.append({
        "path": path,
        "width": width,
        "height": height
        }
    )
```
