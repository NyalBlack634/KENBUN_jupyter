一つのタプルに結果が入っている<br>
必要なのは、パスの部分だけだからそこだけを取り出すような関数にする

```python
from functools import singledispatch

@singledispatch
def serach_index_keyword(keyword: str, cur):
    """見出しテキストとキーワードが部分一致する見出しを取り出す<br>
    keywordが一つの場合

    Args:
        keyword (str): db内を検索するキーワード
        cur (sqlite3.connect('db').cursor): データベースを操作するカーソルオブジェクト

    Returns:
        list[tuple()]: 検索結果
    """

    sql = f'''
    SELECT company, year, month, page FROM headline WHERE ocr_result LIKE '%{keyword}%'
    '''

    cur.execute(sql)
    return cur.fetchall()

@serach_index_keyword.register
def _(keywords: list, cur):
    """見出しテキストとキーワードが部分一致する見出しを取り出す<br>
    keywordが複数の場合(こっちの方が多いだろう)

    Args:
        keywords (list): db内を検索するキーワード
        cur (sqlite3.connect('db').cursor): データベースを操作するカーソルオブジェクト

    Returns:
        _type_: 検索結果
    """
    sql = f'''
    SELECT company, year, month, page FROM headline WHERE
    '''

    for keyword in keywords:
        sql += f" ocr_result LIKE '%{keyword}%' OR"
    else:
        # 最後のORを削除
        sql = sql[::-1] # 文字列反転
        delete_or = ' OR'
        delete_or = delete_or[::-1]
        sql = sql.replace(delete_or, '', 1)
        sql = sql[::-1] # 元の並びに戻す

        cur.execute(sql)
    
    return cur.fetchall()
```

```python
serach_index_keyword('大分', cur)
```

```python
serach_index_keyword(['大分', '戦争'], cur)
```

```python
import os

from PIL import Image

def parse_result(search_results):
    """DBから得た結果を整形

    Args:
        search_results (list(tuple())): DBから得た検索結果

    Returns:
        _type_: _description_
    """
    
    result_list = []

    # 見出し画像を格納しているパス
    PIC_DIR_PATH = 'C:/xampp/htdocs/KENBUN_landmark_Integration_nakano/img/landmark/769_hl_10_trimLandmark/scale0/SNBN/'

    for company, year, month, page in search_results:
        path = PIC_DIR_PATH + '/'.join([company, year, month, page]) + '.jpg'
        if os.path.exists(path):
            img = Image.open(path)
            width, height = img.size  # 画像の幅と高さを取得
        else:  # エラーを避けるために0を代入
            width = height = 0
            
        append_dic = {
            "company_name": company,
            "year": year,
            "month": month,
            "page": page,
            "width": width,
            "height": height
            }
        result_list.append(append_dic)

    return result_list
```

```python
len(parse_result(serach_index_keyword('大分', cur)))
```
