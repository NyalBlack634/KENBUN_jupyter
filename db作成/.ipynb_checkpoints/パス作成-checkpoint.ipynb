```python
from PIL import Image
import os
```

```python
i=0
path_list = []

PIC_DIR_PATH = 'C:/xampp/htdocs/KENBUN_landmark_Integration_nakano/img/landmark/769_hl_10_trimLandmark/scale0/SNBN/'
for midashi_id, company, year, month, page in result:
    path = PIC_DIR_PATH + '/'.join([company, year, month, page]) + '.jpg'
    
    if os.path.exists(path):
        img = Image.open(path)
        width, height = img.size  # 画像の幅と高さを取得
    else:  # エラーを避けるために0を代入
        width = height = 0
    
    # (id, パス、横、高さ)
    path_list.append((midashi_id, '/'.join([company, year, month, page])+'.jpg', width, height))
    
    print('\r'+ f'{str(i)} / {len(result)}', end='')
    i+=1
```

```python
import pickle
```

```python
with open('path_info.pickle', mode='wb') as pk:
    pickle.dump(path_list, pk)
```

```python
path_list
```

```python
# executemany()用
path_info = []
size_info = []

for midashi_id, path, width, height in path_list:
    path_info.append((midashi_id, path))
    size_info.append((midashi_id, width, height))
```

```python
with open('size_info.pickle', mode='rb') as pk:
    size_info = pickle.load(pk)
```

```python
# sqlite3で書く場合
import sqlite3

conn = sqlite3.connect('serach_index.db')
cur = conn.cursor()
```

```python
# テーブル作成
cur.execute("CREATE TABLE headline_paths (id integer, path, PRIMARY KEY (id))")
```

```python
# テーブル作成
cur.execute("CREATE TABLE headline_size (id integer, width integer, height integer, PRIMARY KEY (id))")
```

```python
cur.execute("select * from sqlite_master where type='table';")
```

```python
cur.fetchall()
```

```python
# テーブル名: headline_paths
cur.executemany("INSERT INTO headline_paths (id, path) values (?, ?)", path_info)
```

```python
# テーブル名
cur.executemany("INSERT INTO headline_size (id, width, height) values (?, ?, ?)", size_info)
```

```python
cur.execute("SELECT * FROM headline_paths WHERE id < 10;")
```

```python
cur.fetchall()
```

```python
cur.execute("SELECT * FROM headline_size WHERE id < 10;")
```

```python
cur.fetchall()
```

```python
conn.commit()
```

```python
# キーワードによる検索を行う場合の検索文(過程)
keyword = '大分' # テスト用

sql = f"""
SELECT path, width, height, ocr_result FROM headline
INNER JOIN headline_paths ON headline.id = headline_paths.id
INNER JOIN headline_size ON headline.id = headline_size.id
WHERE ocr_result LIKE '%{keyword}%';
"""
```

```python
cur.execute(sql)
```
