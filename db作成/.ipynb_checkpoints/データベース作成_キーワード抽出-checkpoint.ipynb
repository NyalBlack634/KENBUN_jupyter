```python
import os
```

```python
path_list = []

for folder, sub_folder, files in os.walk('E:/大学/result'):
    for file in files:
        path_list.append(folder + '/' + file)
```

```python
import pickle

keyword_results = []

for path in path_list:
    with open(path, mode='rb') as pf:
        keyword_results.append(pickle.load(pf))
```
