```python
import pickle
import os
```

```python
for _, _, files in os.walk('result'):
    file_list = [file for file in files]
```

```python
file_list
```

```python
import pprint

for n, file in enumerate(file_list):
    if n < 20:
        print('==' * 20)
        with open('result/' + file, mode='rb') as p:
            result = pickle.load(p)
        pprint.pprint(result)
        print(len(result[1]))
```
